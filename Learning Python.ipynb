## Notebooks
Notebooks are similar to scripts in that they contain multiple commands, but offer 4 additional benefits:
1. You can intersperse rich text and formatting in them (this document is a notebook)
2. You can show inline figures and graphs
3. You can break them into segments and run each segment independently and out of order (this should be done carefully, as it can cause problems)
4. You can edit them interactively without needing to rerun everything

In practice, points 3 and 4 obviate the need to use the python command line for most purposes.

## Flow control

### For Loops


For loops are an easy way to iterate across a vector (or simply doing things a lot of ways.) A for loop will go through every value in a vector, for example

````python
for i in ourfirstvector:
    print(i)
````

It's sometimes helpful to do have both what the index is and the value: enumerate helps with this:

````python
for i,j in enumerate(ourfirstvector):
    print(i,j)
````

````python
#You can nest for loops, for example to go through a matrix
for i in ourfirstmatrix:
    for j in i:
      print(j)
````

### If statements
If statements are used to check something. They look for true or false and only do the commands in them if there is a true. Else can be used to get the opposite.

````python
x=True
y=False

if x:
    print('x is true')
else: 
    print('x is false')

if y:
    print('y is true')
else:
    print('y is false')
````

More often than just the true vs false you use comparisons to generate the truth, e.g.


````python
x=5
y=10
z=5
a=1

if x==z:
    print('x and z are equal')

if x>y:
    print('x is greater than y')
elif x<y: #elif is short for else if, allowing to check multiple options
    print('x is less than y')
else:
    print('x is equal to y')
````

Combining for loops and if statements can let you check arbitrarily large amounts of data.

````python
randomnumbers=np.random.rand(100) #generate 100 random numbers between 0 and 1
count_above_half=0
for i in randomnumbers:
    if i>0.5:
        count_above_half=count_above_half+1
print(count_above_half)
````

That said, you can also do elementwise comparisons on vectors, which is often faster and easeir

````python
randomnumbers>0.5
````

````python
np.sum(randomnumbers>0.5) #sum ads up a vector, treats true as 1 and false as zero
````

You can also use true and false as indexes

````python
#show just values above .5
randomnumbers[randomnumbers>0.5]
````

## Melt

In Pandas (this is also true of R, and is a good principle generally), there are two types of datasets:
### Wideform
This is your typical spreadsheet:
| |A|B|C
|--|--|--|--
1|A1|B1|C1
2|A2|B2|C2

This is easily interpretable for data with two variables, but becomes complciated when you have multiple factors (Usually ingvolving multiple *Index* Columns). Instead, it's good practice to be comfortable with "Longform Data





### Longform Data
Longform data

|Index|Variable|Value|
|--|--|--|
1|A|A1
1|B|B1
1|C|C1
2|A|A2
2|B|B2
2|C|C2
